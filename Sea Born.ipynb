```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
dst=pd.read_excel('student.xlsx')
```

```python
dst.head()
```

```python
sns.lineplot(x="id",y="mark",data=dst)
plt.show()
```

```python
sns.lineplot(x="id",y="mark",data=dst,hue='gender')
plt.show()
```

```python
sns.lineplot(x="id",y="mark",data=dst,hue='gender',style='gender')
plt.show()
```

```python
sns.lineplot(x="id",y="mark",data=dst,hue='gender',style='gender',markers=True)
plt.show()
```

```python
sns.set(style='whitegrid')
```

```python
sns.barplot(x='gender',y='mark',data=dst)
plt.show()
```

```python
dst.shape
```

```python

sns.barplot(x='gender',y='mark',data=dst,hue='class',palette='rocket')
plt.show()
```

```python
sns.scatterplot(x="id",y="mark",data=dst,hue='gender')
plt.show()
```

```python
sns.distplot(dst['mark'],color='red')
plt.show()
```

```python
sns.distplot(dst['mark'],hist=False,color='g')
plt.show()
```

```python
sns.distplot(dst['mark'],kde=False,color='g')
plt.show()
```

```python
sns.distplot(dst['mark'],kde=False,color='g',vertical=True)
plt.show()
```

```python
sns.jointplot(x='class',y='mark',data=dst, color='brown')
plt.show()
```

```python
sns.pairplot(dst,hue='class')
plt.show()
```
