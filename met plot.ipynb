```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x= np.arange(1,11)
y=x*3
plt.plot(x,y,color="g",linestyle=':',linewidth=5)
plt.title("single line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
y1=2*x
y2=3*x
plt.plot(x,y1,color='r',linestyle=':',linewidth=3)
plt.plot(x,y2,color='g',linewidth=3)
plt.title("two line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid("true")
plt.show()

```

```python
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.plot(x,y1,color='r',linestyle=':',linewidth=3)
plt.plot(x,y2,color='g',linewidth=3)
plt.title("multiple line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid("true")
plt.subplot(1,2,2)
plt.plot(x,y1,color='r',linestyle=':',linewidth=3)
plt.plot(x,y2,color='g',linewidth=3)
plt.title("multiple line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid("true")
plt.show()
```

```python
Student={"bob":78,"sahil":99,"rahil":100,"basit":101}


```

```python
names=list(Student.keys())
makrs=list(Student.values())
```

```python
plt.bar(names,makrs)
plt.title("barplot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
plt.barh(names,makrs,color="g")
plt.title("barplot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid("true")
plt.show()
```

```python
a=[2,3,4,6,7,8,9,10]
b=[4,5,6,7,8,9,2,11]
plt.scatter(a,b,marker='*',color='red')
plt.title("scatter")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid("true")
plt.show()
```

```python
a=[2,3,4,6,7,8,9,10]
b=[4,5,6,7,8,9,2,11]
c=[6,7,8,7,3,6,1,3]
plt.scatter(a,b,marker='*',color='red',s=100)
plt.scatter(a,c,color='g',s=100)
plt.show()
```

```python
a=[2,3,4,6,7,8,9,10]
b=[4,5,6,7,8,9,2,11]
c=[6,7,8,7,3,6,1,3]
plt.subplot(1,2,1)
plt.scatter(a,b,marker='*',color='red',s=100)
plt.subplot(1,2,2)
plt.scatter(a,c,color='g',s=100)
plt.show()
```

```python
#pie chart
subject=['science','physics','arabic','urdu','math']
marks=[75,66,33,55,92]
plt.pie(marks,labels=subject,autopct='%0.1f%%',colors=['yellow','grey','blue','red','green'])
plt.show()
```
